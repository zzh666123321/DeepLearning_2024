```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('222.jpg',0)
# 
ret1,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY) # 
# Otsu
ret2,th2 = cv.threshold(img,0,255,cv.THRESH_OTSU)
# 
ret3,th3 = cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
plt.subplot(2,2,1)
plt.imshow(img,'gray')
plt.title('origin image')
plt.subplot(2,2,2)
plt.imshow(th1,'gray')
plt.title('BINARY:'+str(ret1))
plt.subplot(2,2,3)
plt.imshow(th2,'gray')
plt.title('OTSU:'+str(ret2),y=-0.4)
plt.subplot(2,2,4)
plt.imshow(th2,'gray')
plt.title('BINARY+OTSU:'+str(ret3),y=-0.4)
plt.show()
```

```python
import cv2.cv2 as cv
import numpy as np
import os
np.seterr(invalid='ignore') #
import matplotlib.pyplot as plt

filepath = './Image'
def otsu3layers(imagepath):  # 
    img = cv.imread(filepath+'/'+imagepath, 0)  # 
    img = img.ravel()
    high, wide = img.shape
    pixels = cv.calcHist([img], [0], None, [256], [0, 256])  # 256,1
    p = pixels / (high * wide)  # p_i
    x=np.linspace(1,256,256) #02551256
    x=x.reshape(256,1)  # 1shape256,1

    maxvar=0
    th1=0
    th2=0
    cnt=0
    for k1 in range(1,257):
        for k2 in range(k1+1,257):
            w0 = np.sum(p[0:k1])
            w1 = np.sum(p[k1:k2])
            w2 = np.sum(p[k2:256])  # pythonnp.sum(p[0:256])=1
            u0 = np.dot(x[0:k1].T, p[0:k1])
            u1 = np.dot(x[k1:k2].T, p[k1:k2])
            u2 = np.dot(x[k2:256].T, p[k2:256])
            W = np.array([[w0, w1, w2]]).T  # np.array,
            U = np.array([[int(u0), int(u1), int(u2)]]).T
            UT = np.sum(U)
            temp = (U - UT * W) * (U - UT * W) / W
            varbetween = np.sum(temp)
            if varbetween>maxvar:
                cnt=cnt+1
                maxvar=varbetween
                th1=k1
                th2=k2
                # print(""+str(cnt)+""+str(th1)+","+str(th2)+""+str(maxvar))
    # mean1 = int(np.dot(x[0:th1].T, p[0:th1]))
    # mean2 = int(np.dot(x[th1:th2].T, p[th1:th2]))
    # mean3 = int(np.dot(x[th2:256].T, p[th2:256]))
    for row in range(0,high):
        for col in range(0,wide):
            if img[row, col]<th1:
                img[row, col]=0
            elif img[row, col]>=th1 and img[row, col]<=th2:
                img[row, col] =0
            elif img[row, col]>th2 and img[row, col]<=230:
                img[row, col] = 255
            else:
                img[row, col] = 0
    path = imagepath.split('.')[0]
    path = path + '.png'
    print(path)
    cv.imwrite('./Mask'+'/'+path,img)
    #plt.imshow(img,'gray')
    #plt.show()
    #cv.waitKey()
    #cv.destroyAllWindows()


if __name__ == '__main__':
    
    dirs = os.listdir(filepath)
    for a in dirs:
        imagepath = filepath + '/' + a
        print(a)
        otsu3layers(a)


    #otsu3layers("222.jpg")
```
