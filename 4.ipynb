```python
color1 = [0,0,0]     #
color2 = [0,0,255]  # 
color3 = [95,158,160]   # 
color4 = [102,205,170]   # 
color5 = [255,0,0]  # 
#
th1 = 5
th2 = 35
th3 = 90
th4 = 120
for i in range(0,5):
    img[i] = cv.GaussianBlur(img[i], (21,21),0)

img1 = np.zeros((5, 350, 300,3),dtype=np.uint8)
for i in range(0,5):
    for row in range(0,h):
        for col in range(0,w):
            if img[i][row][col]<=th1:
                img1[i,row, col]=color1
            elif img[i][row][col]>th1 and img[i][row][col]<=th2:
                img1[i,row, col] =color2
            elif img[i][row][col]>th2 and img[i][row][col]<=th3:
                img1[i,row, col] = color3
            elif img[i][row][col]>th3 and img[i][row][col]<=th4:
                img1[i,row, col] = color4
            elif img[i][row][col]>th4:
                img1[i,row, col] = color5
plt.figure(figsize=(25,20))
for i in range(0,5): 
    plt.subplot(1,5,i+1)
    plt.axis('off')
    plt.title(file[i])
    plt.imshow(img1[i])
plt.show()
```

```python
import cv2 as cv
import numpy as np
np.seterr(invalid='ignore') #
import matplotlib.pyplot as plt
import time
start = time.perf_counter()


def ostu5layers(img,i):  # 
    print('Processing:img-'+str(i))
    high, wide = (h,w)
    pixels = cv.calcHist([img], [0], None, [256], [0, 256])  # 256,1
    p = pixels / (high * wide)  # p_i
    x=np.linspace(1,256,256) #02551256
    x=x.reshape(256,1)  # 1shape256,1

    maxvar=0
    th1=0
    th2=0
    th3=0
    th4=0
    cnt=0
    for k1 in range(1,10):
        for k2 in range(k1+1,50):
            for k3 in range(50,110):
                for k4 in range(110,140):
                    w0 = np.sum(p[0:k1])
                    w1 = np.sum(p[k1:k2])
                    w2 = np.sum(p[k2:k3])
                    w3 = np.sum(p[k3:k4])
                    w4 = np.sum(p[k4:256])  # pythonnp.sum(p[0:256])=1
                    
                    u0 = np.dot(x[0:k1].T, p[0:k1])
                    u1 = np.dot(x[k1:k2].T, p[k1:k2])
                    u2 = np.dot(x[k2:k3].T, p[k2:k3])
                    u3 = np.dot(x[k3:k4].T, p[k3:k4])
                    u4 = np.dot(x[k4:256].T, p[k4:256])
                    
                    W = np.array([[w0, w1, w2, w3, w4]]).T  # np.array,
                    U = np.array([[int(u0), int(u1), int(u2), int(u3), int(u4)]]).T
                    UT = np.sum(U)
                    temp = (U - UT * W) * (U - UT * W) / W
                    varbetween = np.sum(temp)
                    if varbetween>maxvar:
                        cnt=cnt+1
                        maxvar=varbetween
                        th1=k1
                        th2=k2
                        th3=k3
                        th4=k4
                        print(""+str(cnt)+""+str(th1)+","+str(th2)+","+str(th3)+","+str(th4)+","+""+str(maxvar))
                        end = time.perf_counter()
                        print('Running time: %s Seconds' % (end - start))
    # mean1 = int(np.dot(x[0:th1].T, p[0:th1]))
    # mean2 = int(np.dot(x[th1:th2].T, p[th1:th2]))
    # mean3 = int(np.dot(x[th2:256].T, p[th2:256]))
    
    img11 = np.zeros((350, 300,3),dtype=np.uint8)
    for row in range(0,h):
        for col in range(0,w):
            if img[row][col]<=th1:
                img11[row, col]=color1
            elif img[row][col]>th1 and img[row][col]<=th2:
                img11[row, col] =color2
            elif img[row][col]>th2 and img[row][col]<=th3:
                img11[row, col] = color3
            elif img[row][col]>th3 and img[row][col]<=th4:
                img11[row, col] = color4
            elif img[row][col]>th4:
                img11[row, col] = color5
    return img11

if __name__ == '__main__':
    img1 = np.zeros((5, 350, 300,3),dtype=np.uint8)
    for i in range(0,5):
        start = time.perf_counter()
        img1[i]=ostu5layers(img[i],i)
    plt.figure(figsize=(25,20))
    for i in range(0,5): 
        plt.subplot(1,5,i+1)
        plt.axis('off')
        plt.title(file[i])
        plt.imshow(img1[i])
    plt.show()
```
